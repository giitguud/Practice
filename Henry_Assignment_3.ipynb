3. Make an interactive scatter plot, using Altair, showing the correlation between a movie's budget, on one axis, and its worldwide gross income on another. What movies have very low budgets and very high income? (You may want to annotate points with movie titles, or add movie titles to the hover text of points.)
For #3 you'll notice that you can't really plot all 8.25 gajillion movies, as points in a scatter plot, so you'll have to choose a subset. Why not choose movies that are made in the US, between 2000 and 2020, and that have budgets greater than $100,000? That'll give you more meaningful results.

```python
####USA movies 2000-2020, budget >1000000, worldwide gross
```

```python
working_df = movies_df.filter(items=['title','country','year','budget','worlwide_gross_income'])
```

```python
working_df['cleanYear'] = working_df['year'].apply(strToIntAwesome)
```

```python
budget_known = working_df[working_df['budget'].notna()]
```

```python
gross_known = budget_known[budget_known['worlwide_gross_income'].notna()]
```

```python
gross_known
```

```python
USA_only = gross_known[gross_known['country'] == 'USA']
```

```python
def remove_dollar_sign(budget):
    if type(budget) == str and budget.startswith('$'): 
        budget = budget[2:]
    return int(budget)
```

```python
def remove_dollar_sign_gross(worlwide_gross_income):
    if type(worlwide_gross_income) == str and worlwide_gross_income.startswith('$'): 
        worlwide_gross_income = worlwide_gross_income[2:]
    return int(worlwide_gross_income)
```

```python
dollars_only = USA_only[USA_only['budget'].str.contains('\$')]
```

```python
#dollars_only['cleanbudget'] = dollars_only['budget'].apply(remove_dollar_sign)
```

```python
dollars_only['cleangross'] = dollars_only['worlwide_gross_income'].apply(remove_dollar_sign_gross)
```

```python
dollars_only
```

```python
dollars_only.drop(['year'], axis=1)
```

```python
Selected_years = dollars_only[(dollars_only['cleanYear'] >= 2000) & (dollars_only['cleanYear'] <= 2020)]
```

```python
Selected_years_dropped = Selected_years.drop(['country','budget','worlwide_gross_income','year'], axis=1)
```

```python
Over_million = Selected_years_dropped[(Selected_years['cleanbudget'] >= 1000000)]
```

```python
Over_million
```

```python
alt.Chart(Over_million).mark_circle(size=60).encode(
    x='cleanbudget',
    y='cleangross',
    color='cleanYear:O',
    tooltip=['title', 'cleanYear', 'cleanbudget', 'cleangross']
).interactive()
```

```python
just20152020 = Selected_years_dropped[(Selected_years_dropped['cleanYear'] >= 2015) & (dollars_only['cleanYear'] <= 2020)]
just20152020_rich = just20152020[(just20152020['cleanbudget'] >= 1000000)]
```

```python
alt.Chart(just20152020_rich).mark_circle(size=60).encode(
    x='cleanbudget',
    y='cleangross',
    color='cleanYear:O',
    tooltip=['title', 'cleanYear', 'cleanbudget', 'cleangross']
).interactive()
```
